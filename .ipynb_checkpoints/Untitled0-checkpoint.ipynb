```python
fraud = input_data[input_data.Class == 1]
non_fraud = input_data[input_data.Class == 0]
```

```python
train1 = fraud.sample(frac=0.70)
train2 = non_fraud.sample(frac=0.70)
trainDF = pd.concat([train1, train2], axis = 0)
```

```python
testDF = input_data.loc[~input_data.index.isin(trainDF.index)]
```

```python
trainDF.shape
```

```python
testDF.shape
```

```python
trainDF = shuffle(trainDF)
testDF = shuffle(testDF)

trainY = trainDF['Class']
testY = testDF['Class']
```

```python
trainDF = trainDF.drop(['Class'], axis = 1)
testDF = testDF.drop(['Class'], axis = 1)
```

```python
features = trainDF.columns.values

for feature in features:
    mean, std = input_data[feature].mean(), input_data[feature].std()
    trainDF.loc[:, feature] = (trainDF[feature] - mean) / std
    testDF.loc[:, feature] = (testDF[feature] - mean) / std
```
